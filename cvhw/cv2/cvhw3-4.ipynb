```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
image_color = cv.imread("img.jpg")
```

```python
image = cv.cvtColor(image_color,cv.COLOR_BGR2GRAY)
```

```python
hist = cv.calcHist([image],[0],None,[256],[0,256])
```

```python
hist.shape
```

```python
plt.hist(image.ravel(),256);
plt.show()
```

```python
# mask = np.zeros(image.shape[0:2],np.uint8)
eq = cv.equalizeHist(image)
plt.hist(eq.ravel(),256)
plt.show()
```

```python
res = np.hstack((image,eq))
plt.figure()
plt.imshow(res,plt.cm.gray)
```
