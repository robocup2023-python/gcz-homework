```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
image_color = cv.imread("img2.jpeg")
image = cv.cvtColor(image_color,cv.COLOR_BGR2GRAY)
```

```python
image = cv.GaussianBlur(image, (5, 5), 0.4)
```

```python
desentx = cv.Sobel(image,cv.CV_64F,1,0)
desenty = cv.Sobel(image,cv.CV_64F,0,1)
```

```python
desentx_square = np.square(desentx)
desenty_square = np.square(desenty)
desent_theta = np.arctan2(desenty,desentx)
desentxy = desentx * desenty
```

```python
desentx_square = cv.GaussianBlur(desentx_square, (5,5), 0.5)
desenty_square = cv.GaussianBlur(desenty_square, (5,5), 0.5)
desentxy = cv.GaussianBlur(desentxy, (5,5),  0.5)
```

```python
M = np.zeros((*image.shape,2,2))
M[:,:,0,0] = desentx_square
M[:,:,0,1] = M[:,:,1,0] = desentxy
M[:,:,1,1] = desenty_square
```

```python
R = np.linalg.det(M) - 0.06*np.trace(M,axis1=-1,axis2=-2)
```

```python
th = np.max(R)*0.01
mask = R > th
print(mask)
```

```python
def NMS(arr,i,j,dir):
    maxi,maxj = arr.shape
    if dir > 3*np.pi/8 or dir < -3*np.pi/8:
        dx,dy = [1,-1],[0,0]
    elif dir > np.pi/8:
        dx,dy = [1,-1],[1,-1]
    elif dir > -np.pi/8:
        dx,dy = [0,0],[1,-1]
    else:
        dx,dy = [1,-1],[-1,1]
    for k in range(2):
        newx,newy = i+dx[k],j+dy[k]
        if newx > -1 and newx < maxi and newy > -1 and newy < maxj:
            if arr[newx,newy] > arr[i][j]:
                return True
    return False
```

```python
height, width = image.shape
for i in range(height):
        for j in range(width):
            if mask[i,j] and NMS(R,i,j,desent_theta[i,j]):
                R[i,j] = 0
                mask[i,j] = False
img = np.dstack((image,image,image))
img[mask] = [255,0,0]
```

```python
plt.figure()
plt.imshow(img,plt.cm.gray)
```
